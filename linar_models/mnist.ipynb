```python
# Load MNIST Dataset
from tensorflow.keras.datasets import mnist

(X_training, S_training), (X_test, S_test) = mnist.load_data()

print("Shape of X_training: ".ljust(20),  X_training.shape)
print("Shape of S_training: ".ljust(20),  S_training.shape)
print("Shape of X_test: ".ljust(20),  X_test.shape)
```

```python
# Prepare data for linear system modeling
X_training = X_training.reshape(60000,784)
X_test = X_test.reshape(10000,784)

print("Shape of X_training: ".ljust(20),  X_training.shape)
print("Shape of X_test: ".ljust(20),  X_test.shape)
```

```python
from matplotlib import pyplot as plt
import numpy as np

def display_sample(sample: np.ndarray):
    pixels = sample.reshape((28, 28))
    plt.imshow(pixels, cmap='viridis')
    plt.show()

def display_samples(samples: np.ndarray, width = 6):
    number_of_samples = samples.shape[0]
    number_of_rows = int(number_of_samples/width) + (1 if (number_of_samples % width != 0) else 0)
    fig, axs = plt.subplots(number_of_rows, width)
    for i in range(number_of_samples):
        axs[int(i/width), i % width].imshow(X_training[i].reshape((28, 28)), cmap=plt.get_cmap('viridis'))
    fig.set_figwidth(20)
    plt.show()

display_samples(X_training[:12])
```
